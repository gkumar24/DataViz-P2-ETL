```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
csv_path = "Resources\census_data.csv"
census_df = pd.read_csv(csv_path)

census_df.head()
```

```python
#Viewing columns to determine the columns that we will be dropping/Extracting from the dataframe
census_df.columns
```

```python
#Using loc method to choose columns ww will use for analysis
census_demographics = census_df.loc[:, ["State", "County", "TotalPop", "Men", "Women", "Hispanic", "White", "Black", "Native", "Asian", "Pacific"]]
census_demographics
```

```python
#find the date types for the different variables in the table. 
census_demographics.dtypes    
```

```python
#Groupby County and suming by demographics to have one row for each county.
census_dem = census_demographics.groupby(['State','County']).sum()
census_dem.head()
```

```python
#Resetting index so that State and County are set to columns. 
census_dem = census_dem.reset_index()
```

```python
census_dem
```

```python
#Importing SqlAlchemy to use import DataFrame into Postgress.
import sqlalchemy
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
#converting to lower_case to match postgres naming conversion for this particular table
census_dem.columns = [t.lower() for t in census_dem.columns]
```

```python
#Get connection credentials from config file
from config import username, password
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/TargetCensusDB')
# engine = create_engine(f'postgresql+psycopg2://{username}:{password}@localhost:5432/census_data')
connection = engine.connect()
```

```python
#Export Row file to Postgres
engine.execute("delete FROM census_df")
census_df.to_sql('census_df', engine, if_exists='append', index=False)
```

```python
#Exporting tranformed file to Postgres. Appending since table heads already exist in Postgres
engine.execute("delete FROM census_race_ethnicity")
census_dem.to_sql('census_race_ethnicity', engine, if_exists='append', index=False)
```

```python
# Read title table into dataframe
StoreCount_Select = "SELECT state, county, state_abbr, store_count from target_count_county"
targetCount_byCounty = pd.read_sql(StoreCount_Select, connection)
targetCount_byCounty
```

```python
target_census = pd.merge(targetCount_byCounty, census_dem, how='left', on=['county', 'state'])
target_census
```

```python
#Export to Postgres
engine.execute("delete FROM target_race_ethnicity_stats")
target_census.to_sql('target_race_ethnicity_stats', engine, if_exists='append', index=False)
```
