```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs, re
from splinter import Browser
import pandas as pd
```

```python
# Chrome Driver
executable_path = {'executable_path': '/Chromedrv/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#URL for target location, states
url = 'https://www.allstays.com/c/target-locations.htm'
browser.visit(url)
```

```python
html = browser.html
# Create a Beautiful Soup object
soup = bs(html, 'html.parser')
```

```python
StateBody = soup.find('tbody').find_all('a')
StateBody
```

```python
StateLinks = []
StateNames = []
for link in StateBody:
    # Filter out Map link
    if link.text != 'Map':
        #Link data comes with // and this cause the link to open as file, instead of web page. so had to be removed
        StateLinks.append(link['href'].replace('//',''))
        # City name list
        StateNames.append(link.text)
#         print (link['href'],link.text)
StateLinks
# print(StateNames)
```

```python
States_df = pd.DataFrame(StateNames, columns=["State Name"])
States_df["State Link"] = StateLinks
States_df
```

```python
Target_City_df = pd.DataFrame()

for index, row in States_df.iterrows():
    
    print('https://' + row["State Link"])
    url = 'https://' + row["State Link"]
    state_name = row["State Name"]
    browser.visit(url)
    
    html = browser.html
    # Create a Beautiful Soup object
    soup2 = bs(html, 'html.parser')
    
    StoreContent = soup2.find("div", class_="col-md-5").find_all("h4")
    city = [header.text for header in StoreContent]
    city       
    
    city_df = pd.DataFrame(city,columns=["City"])
    city_df
    
    City_Content = soup2.find_all(string=re.compile("ph:"))
    City_Content 
    
    city_df["City_Content"] = City_Content
    
    city_df["State"] = state_name
    
    city_df["City State Zip"] = city_df["City_Content"].str.rsplit("ph:",n=1,expand = True)[0]
    city_df
    
    city_df["City2"] = city_df["City State Zip"].str.rsplit("\s\n",n=1,expand = True)[0]
    city_df
    
    if Target_City_df.empty:
        Target_City_df = city_df
    else:
        Target_City_df = pd.concat([Target_City_df, city_df], ignore_index=True)
```

```python
Target_City_df
```

```python
States_df.to_csv("Resources\AllState_Target_State_Link.csv",header=True,index=False)
```

```python
Target_City_df["City2"] = Target_City_df["City State Zip"]. \
    str.replace(re.compile(r"[0-9]-[0-9]"),""). \
    replace(re.compile(r"[0-9]"),"")

Target_City_df["City2"] = Target_City_df["City2"].str.strip()

Target_City_df["State_Abbr"] = Target_City_df["City2"].str[-2:]
Target_City_df["State_Abbr"] = Target_City_df["State_Abbr"].str.strip()
```

```python
Target_City_df
```

```python
Target_City_Final_df = Target_City_df.drop(['City2','City_Content'], axis=1)
Target_City_Final_df.rename(columns = {'City State Zip':'City_State_Zip'}, inplace = True)
Target_City_Final_df
```

```python
Target_City_df.to_csv("Resources\TargetCityRaw.csv",header=True,index=False)
Target_City_Final_df.to_csv("Resources\TargetCity.csv",header=True,index=False)
```

```python
uscity = pd.read_csv("Resources/uscities.csv")
uscity.rename(columns = {'state_name':'state'}, inplace = True)
uscity.rename(columns = {'county_name':'county'}, inplace = True)
uscity
```

```python
# Read title table into dataframe
StoreCount_Select = "SELECT city, state, state_abbr, count(city_state_zip) as store_count \
                        FROM public.target_city_location \
                        group by city, state, state_abbr"
targetcity = pd.read_sql(StoreCount_Select, connection)
targetcity
```

```python
target_counties = pd.merge(targetcity, uscity, how='left', on=["city", "state"])
target_counties = target_counties.sort_values(by=["county", "state"])
target_counties
```

```python
targetCount_byCounty = target_counties.groupby(["state","county","state_abbr"]).agg(
    {        
        "store_count" : "sum",        
    })
targetCount_byCounty = targetCount_byCounty.reset_index()
targetCount_byCounty
```

```python
engine.execute("delete FROM target_count_county")
targetCount_byCounty.to_sql('target_count_county', engine, if_exists='append', index=False)
```
